## **Apply Scaling Techniques to Numerical Features:**

```python
# Selecting numerical features
num_features = data.select_dtypes(include=['int64', 'float64']).columns

# Standard Scaling
scaler_standard = StandardScaler()
df_standard_scaled = data.copy()
df_standard_scaled[num_features] = scaler_standard.fit_transform(data[num_features])

# Min-Max Scaling
scaler_minmax = MinMaxScaler()
df_minmax_scaled = data.copy()
df_minmax_scaled[num_features] = scaler_minmax.fit_transform(data[num_features])

# Displaying scaled data
print("Standard Scaled Data:")
print(df_standard_scaled.head())
print("\nMin-Max Scaled Data:")
print(df_minmax_scaled.head())
```

Standard Scaling is preferred when you want to standardize features so that they have a mean of 0 and a standard deviation of 1, which can improve the performance of algorithms that assume normally distributed data   

Min-Max Scaling is useful when you need features to be within a specific range, especially when using algorithms like neural networks or K-Nearest Neighbors.

## **Encoding Techniques:**

**Apply One-Hot Encoding to categorical variables with less than 5 categories.**

```python
 # Selecting categorical features
cat_features = data.select_dtypes(include=['object']).columns

# Apply One-Hot Encoding to categorical variables with less than 5 categories
df_one_hot_encoded = pd.get_dummies(data, columns=[col for col in cat_features if data[col].nunique() < 5])

# Displaying the dataset after encoding
print(df_one_hot_encoded.head())
```

**Use Label Encoding for categorical variables with more than 5 categories.**

```python
from sklearn.preprocessing import LabelEncoder

# Apply Label Encoding
label_encoders = {}
for col in [col for col in cat_features if data[col].nunique() >= 5]:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le

# Displaying the dataset after encoding
print(data.head())
```

**One-Hot Encoding:** It avoids introducing an arbitrary ordinal relationship between categories, which is useful for algorithms like linear regression or tree-based models. However, it can increase the dimensionality of the data significantly.  
**Label Encoding:** Efficient in terms of memory and computation, but it introduces an ordinal relationship between categories, which might not be desirable for some algorithms.

## **Feature Engineering:**

**Create at least 2 new features that could be beneficial for the model. Explain the rationale behind your choices**

```python
# Example: Creating a new feature 'age_bin' based on 'age'
data['age_bin'] = pd.cut(data['age'], bins=[0, 30, 50, 100], labels=['Young', 'Middle-aged', 'Old'])

# Example: Creating a new feature 'hours_per_week_bin' based on 'hours_per_week'
data['hours_per_week_bin'] = pd.cut(data['hours_per_week'], bins=[0, 20, 40, 60, 100], labels=['Part-time', 'Full-time', 'Overtime', 'Extensive'])

# Displaying the dataset with new features
print(data.head())
```

**Rationale behind their creation:**

**age_bin**: Categorizes age into groups (Young, Middle-aged, Old), which might be more informative for the model than raw age values.  
**hours_per_week_bin**: Groups the hours worked per week into categories, which could capture differences in work patterns.

**Apply a transformation (e.g., log transformation) to at least one skewed numerical feature and justify your choice.**

```python
# Calculate skewness for each numerical feature
skewness = data[num_features].skew()
print(skewness)
```

Log transformation is applied to reduce the skewness of data, which can help in making the data more normally distributed and improve model performance.Among all the features capital_gain feature has highly skewed data so lets transform it

```python

#Applying log transformation to 'capital_gain'
data['capital_gain_log'] = np.log1p(data['capital_gain'])

# Check distribution of the original and transformed feature
print(data[['capital_gain', 'capital_gain_log']].describe())
```

## **Feature Selection:**

**Use the Isolation Forest algorithm to identify and remove outliers. Discuss how outliers can affect model performance**

```python
from sklearn.ensemble import IsolationForest

# Fit Isolation Forest
iso_forest = IsolationForest(contamination=0.01, random_state=42)
outliers = iso_forest.fit_predict(data[num_features])

# Add the outliers as a feature and filter them out
data['outlier'] = outliers
df_clean = data[data['outlier'] != -1]

# Display the dataset after removing outliers
print(df_clean.head())
```

Outliers can significantly skew the results of some machine learning models, especially those sensitive to extreme values. Removing outliers can improve model accuracy and generalizability.
