**Median for Numerical columns in the Dataset**

```python
mode_volume=data['Volume'].mode()[0]
mode_AvgPrice=data['Avg Price'].mode()[0]
mode_TotalSalesValue=data['Total Sales Value'].mode()[0]
mode_DiscountRate=data['Discount Rate (%)'].mode()[0]
mode_NetSalesValue=data['Net Sales Value'].mode()[0]
print("mode_volume")
print(mode_volume)
print("mode_AvgPrice")
print(median_AvgPrice)
print("mode_TotalSalesValue")
print(mode_TotalSalesValue)
print("mode_DiscountRate")
print(mode_DiscountRate)
print("mode_NetSalesValue")
print(mode_NetSalesValue)
```
