**	Handle missing values appropriately (e.g., imputation, deletion).**

```python
data.info()
```

```python
#imputing the null values
# replacing the null values with mean of the column
lb_mean=data.LB.mean()
data.LB.fillna(lb_mean,inplace=True)
```

```python
#imputing the null values
# replacing the null values with mean of the column
mltv_mean=data.MLTV.mean()
data.MLTV.fillna(mltv_mean,inplace=True)
```

```python
#imputing the null values

# replacing the null values with median of the column
width_median=data.Width.median()
data.Width.fillna(width_median,inplace=True)

```

```python
#imputing the null values

# replacing the null values with mode of the column
tend_mode=data.Tendency.mode()[0]
data.Tendency.fillna(tend_mode,inplace=True)
```

```python
#imputing the null values

# replacing the null values with mode of the column
nsp_mode=data.NSP.mode()[0]
data.NSP.fillna(nsp_mode,inplace=True)
```
